#### Our task is to forecast ask of forecasting daily cases for next 3 months for 2 different business segments . We need to aggregate Aggregating our data  at the monthly level and sum up the case_count column.

```python
x = list(train['application_date'].values)
ls = []
date = []
month = []
year = []
for i in range(len(x)):
    y = x[i].split("-")
    ls.append(y)
for i in range(len(ls)):
    date.append(ls[i][2])
    month.append(ls[i][1])
    year.append(ls[i][0])
train['date'] = date
train['month'] = month
train['year'] = year
train['weekday'] = pd.to_datetime(train['application_date']).dt.weekday
```
