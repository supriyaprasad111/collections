## TAKING INDEPENDENT AND DEPENDENT VARIABLESS

```python
X=data[['satisfactoryLevel','lastEvaluation','numberOfProjects','avgMonthlyHours','timeSpent.company','workAccident','promotionInLast5years','dept','salary']]
X.head()
```

```python
Y=data['left']
Y.head()
```

```python
## SPLITTING THE DATA INTO TEST AND TRAIN
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

## KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
print("KNN Scores",knn.score(x_test,y_test)*100)    
```

```python
y_pred=knn.predict(x_test)
print(y_pred)
```

```python
##Creating confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm1=confusion_matrix(y_test,y_pred)
print("Confusion Matrix is:",cm1,sep='\n')
```

```python
prob1=knn.predict_proba(x_test)
prob1
```

```python
from sklearn import metrics
```

```python
prob_positive=prob1[:,1]
fpr,tpr,threshold=metrics.roc_curve(y_test,prob_positive)    ##falsepositive rate, true positive rate,
roc_auc1=metrics.auc(fpr,tpr)   ##auc--area under the curve
print('Area under the curve:',roc_auc1)
```

```python
##Plotting ROC curve
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr,tpr,'orange',label='AUC=%0.2f' % roc_auc1)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],color='darkblue',linestyle='--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive rate')
plt.show()
```

#### No, accuracy alone is not a good performane evaluation metric for the given data. A model could have high accuracy but it may come up defining True positive class to be False Positive class or vice - versa

#### From, the above model evaluation, it is obvious that Gradient Boosting is having highest Area_under_curve (0.9876) with model accuracy (0.9804) almost comparable to Random forest (0.984). Hence, we can take Gradient Boosting model as predicting model for this dataset.
